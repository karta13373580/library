```python
df_test=pd.read_csv('submit.csv')



s107 = np.ones((len(df_test),), dtype=int)
s107 = pd.DataFrame(s107)

s2 = np.ones((len(df_test),), dtype=int)
s2 = pd.DataFrame(s2)

x = np.ones((len(df_test),), dtype=int)
x = pd.DataFrame(x)

df_test['學年']=s107
df_test['年級']=x
df_test['學期']=s2

for index, row in df_test.iterrows():
    if row['學年']==1:
        s107[index] = 107
        
for index, row in df_test.iterrows():
    if row['學期']==1:
        s2[index] = 2
        
for index, row in df_test.iterrows():
    if row['學期']==1:
        x[index] = 3
        
df_test['學年']=s107
df_test['年級']=x
df_test['學期']=s2
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
X,Y=shuffle(X,Y,random_state=2)
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=215)
```

```python
X_train.shape
```

```python
from xgboost import XGBClassifier
xgbc=XGBClassifier()
xgbc.fit(X_train,y_train)
preds = xgbc.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,preds)
```
